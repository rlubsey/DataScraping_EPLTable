```python
import pandas as pd
```

```python
#import CSV or excel file from Google Drive
csv = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRzqOL0XWZY2shbtlbwODuC3zsUMan8CPCv_p13qvTC91uyCkeQ1itqvNqL_RjJbexgDUPabvz_S8q2/pub?output=csv')
```

```python
csv
```

```python
#assigned imported CSV dataframe to the variable df
df = csv
```

```python
#drop duplicates from dataframe
df = df.drop_duplicates()
```

```python
#drop any unneeded columns from dataframe
df = df.drop(columns='Not_Useful_Column')
```

```python
df
```

```python
#strip extra characters from values
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df
```

```python
#reformat phone number to exclude special characters
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
```

```python
df
```

```python
#reformat phone numbers to be consistent
df['Phone_Number'] =  df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
#remove nulls and not numbers
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
#split culumns based on delimiter into new columns
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand=True)
df
```

```python
#replace characters to help with consistency
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

df
```

```python
#replace unneeded characters with blanks and fill Na values
df = df.replace('N/a','')
df = df.fillna('')
df
```

```python
#for loop to drop column based on criteria
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'N')

df
```

```python
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace=True)

df
```

```python
#reset index
df = df.reset_index(drop=True)
df
```
