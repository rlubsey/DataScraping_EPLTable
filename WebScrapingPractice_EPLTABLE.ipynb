```python
##import BeautifulSoup
from bs4 import BeautifulSoup
##import requests
import requests
```

```python
## add url for data
url = 'https://www.skysports.com/premier-league-table/2022'
page = requests.get(url)
soup =BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
#create a varible for your table data
table = soup.find('table')

```

```python
#create a variable for your table headings
table_heading = table.find_all('th')

```

```python
#loop through the table headings to create a list
headings = [heading.text for heading in table_heading]
print(headings)
```

```python
#delete any unwaned headings using indexing
del headings[-1]
```

```python
print(headings)
```

```python
#import Pandas
import pandas as pd
```

```python
#create your dataframe and add the column headings
df = pd.DataFrame(columns=headings)
df
```

```python
#create a variable for the content of the table
table_body = soup.find_all('tr')
print(table_body)
```

```python
#loop through the contents referencing the html tags it is wrapped in and the using the lenght of the dataframe to add contents to dataframe
for row in table_body[1:]:
  row_data = row.find_all('td')
  team_data = [data.text.strip() for data in row_data[:10]]
  length = len(df)
  df.loc[length] = team_data
  print(team_data)
```
